```python
import matplotlib.pyplot as plt
import PIL.Image as Image
from lib.scrape import fetch_raster_tile
from lib.utils import get_tile_coordinates
```

```python
longitude, latitude = 6.568458, 46.5188203
zoom = 14
x, y = get_tile_coordinates(longitude, latitude, zoom)

# Default server url is [a,b,c].tile.opentopomap.org
downloaded = fetch_raster_tile(x, y, zoom, "./samples/opentopo/")
```

```python
tile = Image.open(downloaded)
fig = plt.figure(figsize=(5, 5))
plt.imshow(tile)
plt.axis("off")
plt.show()
```

```python
from lib.dataset import OSMTileDataset

dataset = OSMTileDataset(
    data_dir="./samples/zoom_9_tiny/",
)

fig, ax = plt.subplots(1, 3, figsize=(10, 3))
ax = ax.flatten()
for i in range(3):
    tile, caption = dataset[i]
    caption = caption.replace(",", ",\n")
    ax[i].imshow(tile)
    ax[i].set_title(caption, fontsize=10)
    ax[i].axis("off")
fig.savefig("osm_tiles.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
